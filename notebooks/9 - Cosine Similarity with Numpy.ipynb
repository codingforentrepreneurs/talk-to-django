```python
def calculate_cosine_metrics(v1, v2):
    dot_product = np.dot(v1, v2)
    magnitude1 = np.linalg.norm(v1)
    magnitude2 = np.linalg.norm(v2)
    cosine_similarity = dot_product / (magnitude1 * magnitude2)
    cosine_distance = 1 - cosine_similarity
    return int(cosine_similarity* 100), int(cosine_distance * 100)


vector1 = np.array([1, 2, 3])  # Along x-axis
vector2 = np.array([0, 1, 3])  # Along y-axis
vector3 = np.array([3, 0, 1])

compare_vector = np.array([1, 1, 1])


print(calculate_cosine_metrics(vector1, compare_vector))
print(calculate_cosine_metrics(vector2, compare_vector))
print(calculate_cosine_metrics(vector3, compare_vector))
```
